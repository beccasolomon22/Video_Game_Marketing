```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Step 1: Load and preprocess the data
data = pd.read_csv('videogames_final.csv')
data 
```

```python
# data['Success_sales'] = np.where(data['Global_Sales']>= 4, 'Good', 'Bad')
```

```python
# Perform any necessary data cleaning and preprocessing here
# Step 2: Feature Engineering
# Extract relevant features and preprocess them if needed
features = data[['NA_Sales', 'EU_Sales']]
target = data['Genre']
```

```python
# Example feature preprocessing (you can modify this based on your requirements)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
# Step 3: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(scaled_features, target, test_size=0.2, random_state=42)
```

```python
# Step 4: Build and train the neural network model
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(4,)))  # Adjust the input shape based on the number of features
```

```python
model.add(Dense(64, activation='relu'))
```

```python
model.add(Dense(10, activation='softmax'))
```

```python
 # Adjust the output size based on the number of genres
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=10, batch_size=32)
```

```python
# Step 5: Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)
print('Test Loss:', loss)
print('Test Accuracy:', accuracy)
```
