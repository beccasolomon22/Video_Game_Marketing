## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
#from google.colab import drive
import pandas as pd

from google.colab import files
uploaded = files.upload()

#  Import and read the charity_data.csv.
import pandas as pd



Demo1 = pd.read_csv('Games_Global_Sales_Rank.csv',  header=0, encoding='unicode_escape')

Demo2 = pd.read_csv('PS4_GamesSales.csv',  header=0, encoding='unicode_escape')

Demo3 = pd.read_csv('Video_Games_Sales_Users_Rating.csv',  header=0, encoding='unicode_escape')

Demo4 = pd.read_csv('XboxOne_GameSales.csv',  header=0, encoding='unicode_escape')
```

```python
Demo1 = Demo1.rename(columns={'Name': 'Game_Title'})
Demo1 = Demo1.drop(columns='Rank')
Demo1
```

```python
Demo2 = Demo2.rename(columns={'North America': 'NA_Sales',
                                     'Europe': 'EU_Sales',
                               'Japan': 'JP_Sales',
                                'Rest of World': 'Other_Sales',
                                'Global':'Global_Sales',
                                'Game': 'Game_Title'})
Demo2
```

```python
Demo3 = Demo3.rename(columns={'Year_of_Release': 'Year', 'Name': 'Game_Title'})
Demo3 = Demo3.drop(columns=['Critic_Score', 'Critic_Count', 'User_Score',
                            'User_Count', 'Developer', 'Rating'])
Demo3
```

```python
Demo4 = Demo4.rename(columns={'North America': 'NA_Sales',
                                     'Europe': 'EU_Sales',
                               'Japan': 'JP_Sales',
                                'Rest of World': 'Other_Sales',
                                'Global':'Global_Sales',
                                'Game': 'Game_Title'})
Demo4 = Demo4.drop(columns='Pos')
Demo4
```

```python
concat_df = pd.concat([Demo1, Demo2, Demo3, Demo4], ignore_index=True)
concat_df
```

```python
dropped_df = concat_df.drop_duplicates()
dropped_df
```

```python
test_df = dropped_df.dropna()
test_df
```

```python
test_df.to_csv('videogames_final.csv', encoding = 'utf-8-sig')
files.download('videogames_final.csv')
```
